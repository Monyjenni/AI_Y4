**bold text**##**Chinese remainder**

```python
# Python 3.6
from functools import reduce
def chinese_remainder(n, a):
    sum = 0
    prod = reduce(lambda a, b: a*b, n)
    print(f"prod {prod}")

    for n_i, a_i in zip(n, a):
        print(f"n_i:{n_i} & a_i:{a_i}")
        p = prod // n_i
        print(f"p=prod//n_i => {p}={prod}//{n_i}")
        sum += a_i * mul_inv(p, n_i) * p

    print(f"sum:{sum} & {sum}%{prod}:{sum%prod}" )
    return sum % prod

def mul_inv(a, b):
    """
        Multiplactive Inverse under (mod n)
        References:
        - https://byjus.com/maths/multiplicative-inverse/
        - https://www.youtube.com/watch?v=YwaQ4m1eHQo
        - https://www.youtube.com/watch?v=lq285DDdmtw
    """
    print(f"Input a:{a} & b:{b}")
    b0 = b
    x0, x1 = 0, 1
    if b == 1: return 1
    while a > 1:
        q = a // b
        a, b = b, a%b
        x0, x1 = x1 - q * x0, x0
        print(f"q:{q} & a:{a} & b:{b}")
        print(f"x0:{x0} & x1:{x1}")
    if x1 < 0: x1 += b0
    return x1
```
